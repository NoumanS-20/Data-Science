## Task 1: Employee Record
**Columns:** EmpID, Name, Designation, Yrs of Experience, Age, Salary

**Operations:**
1. If Salary > 10000 → drop it
2. Handle NA values in Designation
3. Rename 'Yrs of Experience' as 'exp'
4. Increment Age by 3


```python
import pandas as pd
import numpy as np

# ---------- Create Employee DataFrame ----------
data = {
    'EmpID':      [101, 102, 103, 104, 105],
    'Name':       ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Designation':['Manager', np.nan, 'Analyst', np.nan, 'Developer'],
    'Yrs of Experience': [5, 3, 8, 2, 6],
    'Age':        [28, 25, 35, 22, 30],
    'Salary':     [9000, 12000, 8500, 11000, 9500]
}

df = pd.DataFrame(data)
print('Original DataFrame:')
print(df)
```

```python
# ① Drop rows where Salary > 10000
df = df[df['Salary'] <= 10000]
print('After dropping Salary > 10000:')
print(df)
```

```python
# ② Handle NA values in Designation (fill with 'Unknown')
df['Designation'] = df['Designation'].fillna('Unknown')
print('After handling NA in Designation:')
print(df)
```

```python
# ③ Rename 'Yrs of Experience' as 'exp'
df = df.rename(columns={'Yrs of Experience': 'exp'})
print('After renaming column:')
print(df)
```

```python
# ④ Increment Age by 3
df['Age'] = df['Age'] + 3
print('After incrementing Age by 3:')
print(df)
```

## Task 2: Student Database — 2 DataFrames
**Operations:** Merge (inner, outer, left, right), Join, Concat


```python
# ---------- Create two Student DataFrames ----------
df1 = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name':      ['Arjun', 'Priya', 'Karan', 'Sneha'],
    'Marks':     [85, 90, 78, 92]
})

df2 = pd.DataFrame({
    'StudentID': [3, 4, 5, 6],
    'Name':      ['Karan', 'Sneha', 'Ravi', 'Meena'],
    'Grade':     ['B', 'A', 'C', 'A']
})

print('DataFrame 1:'); print(df1)
print('\nDataFrame 2:'); print(df2)
```

```python
# MERGE — Inner Join (only matching StudentIDs)
inner = pd.merge(df1, df2, on='StudentID', how='inner')
print('Inner Merge:'); print(inner)
```

```python
# MERGE — Outer Join (all records, NaN where no match)
outer = pd.merge(df1, df2, on='StudentID', how='outer')
print('Outer Merge:'); print(outer)
```

```python
# MERGE — Left Join (all from df1, match from df2)
left = pd.merge(df1, df2, on='StudentID', how='left')
print('Left Merge:'); print(left)
```

```python
# MERGE — Right Join (all from df2, match from df1)
right = pd.merge(df1, df2, on='StudentID', how='right')
print('Right Merge:'); print(right)
```

```python
# JOIN (index-based)
df1_idx = df1.set_index('StudentID')
df2_idx = df2.set_index('StudentID')
joined = df1_idx.join(df2_idx, how='inner', lsuffix='_df1', rsuffix='_df2')
print('Join (index-based, inner):'); print(joined)
```

```python
# CONCAT — Stack rows (axis=0)
concat_rows = pd.concat([df1, df2], axis=0, ignore_index=True)
print('Concat (rows, axis=0):'); print(concat_rows)

# CONCAT — Stack columns (axis=1)
concat_cols = pd.concat([df1, df2], axis=1)
print('\nConcat (columns, axis=1):'); print(concat_cols)
```
